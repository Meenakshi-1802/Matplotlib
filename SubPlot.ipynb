```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y,color="indigo")
plt.pie([1],colors="cyan")
x1 = [10,20,30,40,50]
plt.pie(x)
x2 = ["a","b","c","f"]
y2 = [20,40,60,80]
plt.bar(x2,y2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.subplot(2,2,1)
plt.plot(x,y,color="indigo")
plt.subplot(2,2,2)
plt.pie([1],colors="cyan")
x1 = [10,20,30,40,50]
plt.subplot(2,2,3)
plt.pie(x)
x2 = ["a","b","c","f"]
y2 = [20,40,60,80]
plt.subplot(2,2,4)
plt.bar(x2,y2)
plt.show()
```
