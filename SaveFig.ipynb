SaveFig

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="indigo")
plt.savefig("line")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="black")
plt.savefig("line1",dpi=2000)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,6,4]
plt.plot(x,y,color="black")
plt.savefig("line1",dpi=2000,facecolor="yellow")
plt.show()
```
