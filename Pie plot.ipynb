Pie Plot

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
y = ["c","c++","java","python"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
y = ["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

Clockwise Rotation

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
y = ["c","c++","java","python"]
ex = [0.0,0.0,0.4,0.0]
c = ["red","blue","yellow","green"]
plt.pie(x,labels=y,explode=ex,colors = c,autopct="%0.1f%%",shadow = True,radius=1,startangle=180
       ,textprops={"fontsize":13},counterclock=False)
plt.show()
```

linewidth

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
y = ["c","c++","java","python"]
ex = [0.0,0.0,0.4,0.0]
c = ["red","blue","yellow","green"]
plt.pie(x,labels=y,explode=ex,colors = c,autopct="%0.1f%%",shadow = True,radius=1,startangle=180
       ,textprops={"fontsize":13},counterclock=False,wedgeprops = {"linewidth":5,"width":2})
plt.show()
```

Title

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
y = ["c","c++","java","python"]
ex = [0.0,0.0,0.0,0.0]
c = ["red","blue","yellow","green"]
plt.pie(x,labels=y,explode=ex,colors = c,autopct="%0.1f%%",shadow = True,radius=1,startangle=180
       ,textprops={"fontsize":13},counterclock=False,wedgeprops = {"linewidth":5})
plt.title("Tech Languages")
plt.legend(loc=2)
plt.show()
```

Donut pie chart or Multiple pie chart

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
x1 = [40,30,20,10]
y = ["c","c++","java","python"]
c = ["yellow","brown","pink","violet"]
plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.5,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [20,40,60,80]
x1 = [40,30,20,10]
y = ["c","c++","java","python"]
c = ["yellow","brown","pink","violet"]
plt.pie(x,labels=y,radius=1)
plt.pie([1],radius=0.5,colors="white")
plt.show()
```
