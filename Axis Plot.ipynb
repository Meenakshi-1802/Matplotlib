Axis Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,3,1,5,6]
plt.plot(x,y)
plt.xticks(x,labels=["python","c","c++","php","java","R"])
plt.yticks(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,3,1,5,6]
plt.plot(x,y)
plt.xticks(x,labels=["python","c","c++","php","java","R"])
plt.yticks(x)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
