Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,4,5,3,1,6]
plt.scatter(day,no,color="red")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,4,5,3,1,6]
colors=["red","yellow","green","brown","blue","black","orange"]
sizes=[100,200,440,500,300,270,190]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,4,5,3,1,6]
colors=["red","yellow","green","brown","blue","black","orange"]
sizes=[100,200,440,500,300,270,190]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g")
x=plt.colorbar()
x.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.show()
```
