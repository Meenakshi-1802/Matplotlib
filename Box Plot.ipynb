Box Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()

```

Notch Plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,notch = True)
plt.show()
```

Horizontal

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,vert=False)
plt.show()
```

Widths

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,widths=0.5)
plt.show()
```

Labels

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,tick_labels=["Python"])
plt.show()
```

Add colour

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,tick_labels=["Python"],patch_artist=True)
plt.show()
```

Maan

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,tick_labels=["Python"],showmeans=True)
plt.show()
```

Outlier

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,tick_labels=["Python"],showmeans=True)
plt.show()
```

Whisk

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,tick_labels=["Python"],showmeans=True,whis=2)
plt.show()
```

Change Outlier

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,tick_labels=["Python"],showmeans=True,sym="r+",boxprops=dict(color="g")
            ,capprops=dict(color="b"))
plt.show()
```

Whiskerprop

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
plt.boxplot(x,tick_labels=["Python"],showmeans=True,boxprops=dict(color="g")
            ,capprops=dict(color="b"),whiskerprops=dict(color="y"))
plt.show()
```

Filerprops

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,tick_labels=["Python"],showmeans=True,boxprops=dict(color="g") ,capprops=dict(color="b"),whiskerprops=dict(color="y"),flierprops=dict(markeredgecolor="r"))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 =[10,20,60,40,80,70,90]
y = [x,x1]
plt.boxplot(y,tick_labels=["Python","c++"],showmeans=True,boxprops=dict(color="g") ,capprops=dict(color="b"),whiskerprops=dict(color="y"))
plt.show()
```
