```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,3,6]
plt.plot(x,y)
plt.title("Sales",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.text(2,5,"Price",fontsize=15,style="italic")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,3,6]
plt.plot(x,y)
plt.title("Sales",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.text(2,5,"Price",fontsize=15,style="italic",bbox={"facecolor":"cyan"})
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,3,6]
plt.plot(x,y)
plt.title("Sales",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.text(2,5,"Price",fontsize=15,style="italic",bbox={"facecolor":"cyan"})
plt.annotate("Downfall",xy=(2,1),xytext=(4,4),arrowprops=dict(facecolor="black",shrink=90))
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,3,6]
plt.plot(x,y)
plt.title("Sales",fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Growth",fontsize=15)
plt.legend(["up"],loc=1)
plt.show()
```
