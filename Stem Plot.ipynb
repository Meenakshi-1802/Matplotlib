Stem Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,4]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,4]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,4]
plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,4]
plt.stem(x,y,linefmt=":",markerfmt="ro",basefmt="g",label="python")
plt.legend()
plt.show()
```
