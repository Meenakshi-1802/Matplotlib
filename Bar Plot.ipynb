Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","r","c","c++","php","java"]
y=[95,87,76,65,50,88]
plt.xlabel("language",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.title("Bar Plot")
z=["violet","orange","blue","brown","red","green"]
plt.bar(x,y,width=0.5,color=z,align="center",edgecolor="black")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["apple","banana","blueberry","guava","orange","coconut"]
y=[95,87,76,65,50,88]
plt.xlabel("colours",fontsize=20)
plt.ylabel("price",fontsize=20)
plt.title("Fruits")
z=["red","yellow","blue","green","orange","brown"]
plt.bar(x,y,width=0.5,color=z,edgecolor="black",linewidth=3,linestyle=":",alpha=0.6,label="popularity")
plt.legend()
plt.show()
```
