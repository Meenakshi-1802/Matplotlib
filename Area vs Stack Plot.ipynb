Area vs Stack Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area = [2,3,2,5,4]
plt.stackplot(x,area)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [1,3,2,4,2]
area2 = [2,3,4,2,5]
area3 = [1,3,2,4,2]
l = ["are1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l)
plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [1,3,2,4,2]
area2 = [2,3,4,2,5]
area3 = [1,3,2,4,2]
l = ["are1","area2","area3"]

plt.stackplot(x,area1,area2,area3,labels = l,colors = ["cyan","green","yellow"])
plt.legend(loc = 2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [1,3,2,4,2]
area2 = [2,3,4,2,5]
area3 = [1,3,2,4,2]
l = ["are1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l,colors = ["magenta","green","yellow"],baseline="sym")
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc = 2)
plt.show()
```
