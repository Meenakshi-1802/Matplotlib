Bar plot using numpy

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
width=0.5
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Jupyter",fontsize=15)
plt.bar(p,y,width,color="blue",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Jupyter",fontsize=15)
plt.barh(x,y,color="red",label="popularity")
plt.bar(x,z,color="green",label="popularity1")
plt.legend()
plt.show()
```
