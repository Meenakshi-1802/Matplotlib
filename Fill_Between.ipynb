Fill_Between Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.plot(x,y,color="orange")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.plot(x,y,color="orange")
plt.fill_between(x,y)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.plot(x,y,color="orange")
plt.fill_between(x=[2,5],y1=2,y2=5,color="indigo")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
